```python
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import wilcoxon
from statsmodels.stats.proportion import proportions_ztest
```

```python
df_IIvsALS = pd.read_csv('../data/IIvALS.csv')
df_IIvsUU = pd.read_csv('../data/IIvUU.csv')
df_ALSvsUU = pd.read_csv('../data/ALSvUU.csv')
dataframes = [df_IIvsALS, df_IIvsUU, df_ALSvsUU]
dataframe_names = ['IIvALS', 'IIvUU', 'ALSvUU']
fixed_order = ['Much more A than B', 'Bit more A than B', 'About the same', 'Bit more B than A', 'Much more B than A']
```

```python
first_impression_df_with_partial_n = pd.DataFrame(columns=['Condition', 'N Total', 'N Tendency', 'Tendency A', 'Tendency B', '% Tendency B', 'p'])

for dataframe, dataframe_name in zip(dataframes, dataframe_names):
    
    df_tendency = dataframe[dataframe['answer_to_1'].isin(['Much more A than B', 'Bit more A than B', 'Much more B than A', 'Bit more B than A'])]
    
    # Calculate the values for each row
    n_total = len(dataframe)
    n_tendency = len(df_tendency)
    tendency_a = dataframe['answer_to_1'].isin(['Much more A than B', 'Bit more A than B']).sum()
    tendency_b = dataframe['answer_to_1'].isin(['Much more B than A', 'Bit more B than A']).sum()
    percent_tendency_b = (tendency_b / n_total) * 100
    
    # Perform a two-sided proportion test
    count = tendency_a
    nobs = n_tendency
    value = 0.5  # null hypothesis: a/b = 0.5
    _, p_value = proportions_ztest(count, nobs, value)
    
    # Append the values to the result DataFrame
    first_impression_df_with_partial_n = pd.concat([first_impression_df_with_partial_n, pd.DataFrame([{'Condition': dataframe_name, 'N Total': n_total, 'N Tendency': n_tendency, 'Tendency A': tendency_a, 'Tendency B': tendency_b, '% Tendency B': percent_tendency_b, 'p': p_value}])], ignore_index=True)

first_impression_df_with_partial_n
```

```python
def map_answers_to_numbers(column):
    wilcoxon_list = []
    for value in column:
        if value == 'Much more A than B':
            wilcoxon_list.append(-2)
        elif value == 'Bit more A than B':
            wilcoxon_list.append(-1)
        elif value == 'Bit more B than A':
            wilcoxon_list.append(1)
        elif value == 'Much more B than A':
            wilcoxon_list.append(2)
        else:
            pass
    return wilcoxon_list
```

```python
def no_data_check(choice, values):
    return values[choice] == 0
```

```python
for dataframe, dataframe_name in zip(dataframes, dataframe_names):
    for column in dataframe.columns:
        if column.startswith('answer'):
            print(f"Plot for Question: {column} in Dataframe {dataframe_name}")
            plt.figure(figsize=(10, 6))
            
            values_counts = dataframe[column].value_counts().reindex(fixed_order).fillna(0)
                
            res = wilcoxon(map_answers_to_numbers(dataframe[column]))
            
            bar_colour = 'deepskyblue'
            if res.pvalue < 0.01:
                bar_colour = 'navy'
            
            print(f"P-value: {res.pvalue}, Bar Color: {bar_colour}")
            
            ax = values_counts.plot(kind='bar', color=bar_colour, edgecolor='white', width=0.9)
            
            # Iterate over categories and add a line if no data is present
            for i, category in enumerate(fixed_order):
                if no_data_check(category, values_counts):
                    ax.vlines(x=i, ymin=-2, ymax=2, color=bar_colour, linewidth=10)
            
            ax.set_xticks([])
            ax.xaxis.label.set_visible(False)
            ax.set_yticks([])
            
            for spine in ax.spines.values():
                spine.set_visible(False)
            
            plt.show()
```
