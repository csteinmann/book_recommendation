```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_rmse = pd.read_csv('../data/RMSE.csv')

print(df_rmse)
ax = df_rmse.plot(kind='barh', legend=False, y='RMSE Score', x='Algorithm')
ax.set_xlabel('RMSE score')
ax.set_ylabel('')
ax.grid(True)
ax.set_axisbelow(True)
```
