```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_rmse = pd.read_csv('../data/RMSE.csv')

ax = df_rmse.plot(kind='bar', legend=False, y='RMSE Score', x='Algorithm', width=0.6, rot=45)
ax.set_ylabel('RMSE score')
ax.set_xlabel('')
ax.set_title('Accuracy', fontweight='bold')
ax.grid(True)
ax.set_axisbelow(True)
```

```python
df_mean_pop_rank = pd.read_csv('../data/MeanPopRank.csv')

ax = df_mean_pop_rank.plot(kind='bar', legend=False, y='Mean Pop Rank Score', x='Algorithm', width=0.6, rot=45)
ax.set_ylabel('Mean popularity rank')
ax.set_xlabel('')
ax.set_title('Novelty', fontweight='bold')
ax.grid(True)
ax.set_axisbelow(True)
```
